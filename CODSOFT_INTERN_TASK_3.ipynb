```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
x=details.iloc[:,:-1]
y=details['Class']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.35)
```

```python
details.drop_duplicates()
```

```python
details['Class'].value_counts()
```

```python
clf= linear_model.LogisticRegression(C=1e5)
clf.fit(x_train,y_train)
```

```python
y_pred=np. array (clf.predict(x_test))
y=np.array(y_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
print(confusion_matrix(y_test,y_pred))
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
ax= sns.countplot(data=details,x="Class",width = 0.3)
ax.set_title("Class Distribution")
ax.set_xlabel("Class")
ax.set_ylabel("Count")
for i in ax.containers:
  ax.bar_label(i)
plt.show()
```

```python
plt.figure(figsize=(20,12))
ax=sns.heatmap(details.corr(),annot=True,fmt='.2f')
for i in range(details.shape[1]+1):
  ax.axvline(i,linestyle='--',color='black')
  ax.axhline(i,linestyle='--',color='black')
plt.show()
```

```python
print (classification_report(y_test,y_pred))
```

```python
LABLES=['Normal','Fraud']
cm=confusion_matrix(y_test,y_pred)
plt.figure(figsize=(12,12))
sns.heatmap(cm,xticklabels=LABLES,yticklabels=LABLES,annot=True,fmt='g')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
